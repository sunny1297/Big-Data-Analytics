```python
import xml.etree.cElementTree as et
import pandas as pd

def getvalueofnode(node):
    """ return node text or None """
    return node.text if node is not None else None


def main_comments():
    """ main """
    parsed_xml = et.parse("/Users/mandeep/BigData/Project/categories/categories_0001_abcat0010000_to_pcmcat99300050000.xml")
    dfcols = ['category', 'id', 'name']
    df_xml = pd.DataFrame(columns=dfcols)

    for node in parsed_xml.getroot():
        category = node.attrib.get('category')
        id = node.find('id')
        name = node.find('name')

        df_xml = df_xml.append(
            pd.Series([getvalueofnode(category), getvalueofnode(id), getvalueofnode(name)], index=dfcols),
            ignore_index=True)

    return df_xml

category_df =  main_comments()
category_df = category_df.drop('category',axis = 1)
category_df.to_csv('category_df.csv')
```

# XML to CSV Reviews Data

```python
import xml.etree.cElementTree as et
import pandas as pd

def getvalueofnode(node):
    """ return node text or None """
    return node.text if node is not None else None


def main_reviews(filename):
    """ main """
    parsed_xml = et.parse(filename)
    dfcols = ['review', 'id', 'sku','reviewer','name','aboutMe','rating','title','comment','submissionTime']
    df_xml = pd.DataFrame(columns=dfcols)

    for node in parsed_xml.getroot():
        review = node.attrib.get('review')
        id = node.find('id')
        sku = node.find('sku')
        reviewer = node.find('reviewer')
        name = node.find('name')
        aboutMe = node.find('aboutMe')
        rating = node.find('rating')
        title = node.find('title')
        comment = node.find('comment')
        submissionTime = node.find('submissionTime')
        
        df_xml = df_xml.append(
            pd.Series([getvalueofnode(review), getvalueofnode(id), getvalueofnode(sku),
                       getvalueofnode(reviewer), getvalueofnode(name), getvalueofnode(aboutMe),
                       getvalueofnode(rating), getvalueofnode(title), getvalueofnode(comment),
                       getvalueofnode(submissionTime)
                      ], index=dfcols),
            ignore_index=True)

    return df_xml


```

# XML TO CSV USING SPARK

```python
import xml.etree.ElementTree as ElementTree
```

```python
def getproducts(s):
   filetree = ElementTree.fromstring(s)
   return filetree.getiterator('product')
def getsku(product):
    return product.find('sku').text 
def getname(product):
    return product.find('name').text 
def getproductid(product):
    return product.find('productId').text 

# Read XML files into an RDD 

files_products="{PATH}*.xml"
productFiles = sc.wholeTextFiles(files_products)

# Parse each file (as a string) into a collection of product XML records
productRecords = productFiles.flatMap(lambda (filename,xmlstring): getproducts(xmlstring))

# Map each product record to sku, productid and name 
models = productRecords.map(lambda record: getsku(record) + "," +  getname(record)+ "," + getproductid(record))

# Save the data to a file
models.saveAsTextFile("{PATH to save}")
```
